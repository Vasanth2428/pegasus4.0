```python
from src.detector import TrafficViolationDetector
import cv2

# Initialize detector
detector = TrafficViolationDetector()

# Test with video
cap = cv2.VideoCapture('data/test_video.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break
    
    # Process frame
    result_frame, new_events, telemetry = detector.process_frame(frame)
    violation_count = len(new_events)
    # Display results
    cv2.putText(result_frame, f"Violations: {violation_count}", (10, 60),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    cv2.imshow('Traffic Violation Detection', result_frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

# Show violation log
print("Violation Log:")
# Change this:
# for violation in detector.violation_log:
#     print(f"{violation['timestamp']}: {violation['type']} - Duration: {violation.get('duration', 'N/A')}")

# To this:
for violation in detector.violation_log:
    v_type = violation['event_type']
    v_status = violation['metadata']['status']
    v_time = violation['timestamp']
    v_dur = violation.get('duration_seconds', 0)
    print(f"{v_time}: {v_type} ({v_status}) - Duration: {v_dur}s")
```
